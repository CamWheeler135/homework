### Dictionaries and DataFrames

```python
person = {
    "first": "Cameron",
    "last": "Wheeler", 
    "email": "Cameron@email.com"
}
```

When storing data, we know of using dictionaires, above is the data for a single person. But what about multiple? The way to think about pandas is the make every value in the dictioanry a list!!!!

```python
person = {
    "first": ["Cameron"],
    "last": ["Wheeler"], 
    "email": ["Cameron@email.com"]
}
```

So we have a similar dictionary to the first one but we can add more first names and information, so the first value of our list will be the first person! Lets look at multiple people.

```python
people = {
    "first": ["Cameron", "Charlie", "Tom"],
    "last": ["Wheeler", "McGowan", "Padley"], 
    "email": ["Cameron@email.com", "Charlie@email.com", "Tom@email.com"]
}
```

In this dictionary we have the keys with a each value being a list of information that correspond to a person, so if we want to get the information on Cameron we go to the first value in the list, or if we want Toms information we go to the last person on in list. The way to think about this is that the ROWS are the VALUES and the KEYS are the COLUMNS. So they key for email will be the email column and that will contain all of the email values.

```python
people['email']
```

So we can see the values of the key emails, this is similar to how dataframes work, if we want to see the values in the dataframe we can acess the column, just like we have acessed the values of the key. 

```python
# Lets make a dataframe of the dictionary above
practice_df = pd.DataFrame(people)

practice_df
```

Here we have a dataframe of rows and columns, the numbers are called an 'index' this is a unique value for the rows of the dataframe. 

Lets access a single row and column of the dataframe

```python
# Getting the email column, this is returned as a 'series'
practice_df['email']
```

```python
# We might see people accessing a column like this 
practice_df.email
```

It is better to use brackets, as a column could be named the same as something in our program that is not related to the dataframe, as well as this there might be a column in our df that is the same some of the methods that we can use on dataframes such as count. So it is better to just use brackets when working with dataframes to avoid any confusion, but it is good to know that the dot notation exists as other people could use it in projects.

```python
# A series is a row of data, so a row of a single column.
# Think of a dataframe as lots of series, sort of acting as a container for a bunch of series (colums)  b
type(practice_df['email'])
```

```python
# Lets access multiple columns
# We do this by passing a list!!!! we cannot forget to pass a list as it will throw an error
practice_df[['last', 'email']]
```

It is also important to remember that this is not a series as is it not just a row of a column data, this is just a smaller dataframe than the original dataframe!!!!!! 

```python
# Lets get all of the columns in the dataframe, this returns a list of the columns in our dataframe
practice_df.columns
```

Now lets look at getting the rows of the dataframe, to do this we can use 

- loc 
- iloc

We are going to look at iloc first, this allows us to get rows by integer location

```python
practice_df.iloc[0]
```

This shows us the first row of information (remeber that python starts with 0!!!), so we can get a series of the first row of data which is the first, last and email of the person in the frist row which is me!
