### Dictionaries and DataFrames

```python
person = {
    "first": "Cameron",
    "last": "Wheeler", 
    "email": "Cameron@email.com"
}
```

When storing data, we know of using dictionaires, above is the data for a single person. But what about multiple? The way to think about pandas is the make every value in the dictioanry a list!!!!

```python
person = {
    "first": ["Cameron"],
    "last": ["Wheeler"], 
    "email": ["Cameron@email.com"]
}
```

So we have a similar dictionary to the first one but we can add more first names and information, so the first value of our list will be the first person! Lets look at multiple people.

```python
people = {
    "first": ["Cameron", "Charlie", "Tom"],
    "last": ["Wheeler", "McGowan", "Padley"], 
    "email": ["Cameron@email.com", "Charlie@email.com", "Tom@email.com"]
}
```

In this dictionary we have the keys with a each value being a list of information that correspond to a person, so if we want to get the information on Cameron we go to the first value in the list, or if we want Toms information we go to the last person on in list. The way to think about this is that the ROWS are the VALUES and the KEYS are the COLUMNS. So they key for email will be the email column and that will contain all of the email values.

```python
people['email']
```

So we can see the values of the key emails, this is similar to how dataframes work, if we want to see the values in the dataframe we can acess the column, just like we have acessed the values of the key. 

```python
# Lets make a dataframe of the dictionary above
practice_df = pd.DataFrame(people)

practice_df
```

Here we have a dataframe of rows and columns, the numbers are called an 'index' this is a unique value for the rows of the dataframe. 

Lets access a single row and column of the dataframe

```python
# Getting the email column, this is returned as a 'series'
practice_df['email']
```

```python
# We might see people accessing a column like this 
practice_df.email
```

It is better to use brackets, as a column could be named the same as something in our program that is not related to the dataframe, as well as this there might be a column in our df that is the same some of the methods that we can use on dataframes such as count. So it is better to just use brackets when working with dataframes to avoid any confusion, but it is good to know that the dot notation exists as other people could use it in projects.

```python
# A series is a row of data, so a row of a single column.
# Think of a dataframe as lots of series, sort of acting as a container for a bunch of series (colums)  b
type(practice_df['email'])
```

```python
# Lets access multiple columns
# We do this by passing a list!!!! we cannot forget to pass a list as it will throw an error
practice_df[['last', 'email']]
```

It is also important to remember that this is not a series as is it not just a row of a column data, this is just a smaller dataframe than the original dataframe!!!!!! 

```python
# Lets get all of the columns in the dataframe, this returns a list of the columns in our dataframe
practice_df.columns
```

Now lets look at getting the rows of the dataframe, to do this we can use 

- loc 
- iloc

We are going to look at iloc first, this allows us to get rows by integer location

```python
practice_df.iloc[0]
```

This shows us the first row of information (remeber that python starts with 0!!!), so we can get a series of the first row of data which is the first, last and email of the person in the frist row which is me! Also note that the index is the column name now not like the series that is returned when looking at a column, this is becuase if it returned just 0,1,2 etc we would not know what column each value belongs too, so it returns the name of the column as the index.

We can also get specific rows and columns together, the way to think about it is as iloc and loc as functions, and the first argument being the rows and the second argument being the columns, remeber if we just want to columns we can simply use x['y'] to retrieve a specific column but if we want to get both, iloc and loc are used.

In the example below we are collecting the email address for the first two rows using iloc, now when using iloc it is important to remeber that it stands for integer location, so we cannot pass column names like ['email'] as this will not work, we need to use integers. 

```python
# Lets just get the emails for Charlie and Tom
practice_df.iloc[[1,2], [2]]
```

We are now going to look at loc now, which uses index labels much like when calling a specific column, however, becuase the index of our dataframe is just numbers this is going to look alot like iloc at the moment, but when working with the stack overflow data, look for when we are using loc to get specific rows of data.

# Chapter 4 - Filtering, Using Conditional Tests To Filter Out Data

### Using the simple dataset

```python
practice_df
```

```python
# Lets see how many people are called Wheeler in this df, yes its only one, but we are learning!
# We cant use filter as the variable as that is an actual df method
filt = (practice_df['last'] == 'Wheeler')
```

We can see that it returns a boolean value indicating whether the last name is Wheeler or not. This works by looking at if the value matched our filter criteria or not. Note that what is being returned is a series!!!

Now lets create a dataframe using that filter, to do this we just simply pass in the filt varaible into the [] brackets

```python
practice_df[filt]
```

So here we have made a new dataframe with all of the rows that had the last name "wheeler"

But we have defined the filter before placing it into the df, but we can also do it anothe way, where we simply place the whole filter definition into the df instead of assigning it to a variable

```python
practice_df[practice_df['last'] == 'Wheeler']
```

So here we get the same results, but instead of defining the filter first and placing the variable in we simply just place the filter into the square brackets.

On the tutorial, there is also a 3rd way, where we use the loc function, this also allows us to get certain columns as well, see below!

```python
# Getting the email value of the dataframe after the filter is applied
practice_df.loc[filt, 'email']
```

We can also use & for and or | for or to further specify our filtering

```python
# assinging a new filter
filt = (practice_df['last'] == 'Wheeler') | (practice_df['first'] == 'Tom')
```

```python
practice_df[filt]
```

Here we have filtered out for the last name to equal wheeler or the first name equals tom, this returns a df of both rows that match those filters. Note that we have to add the two filter types into their own () 


We can also do the opposite, so return the values that are not caught in this filter, this could be handy when we are cleaning data, for processing!!! See how it is done below.

```python
practice_df[~filt]
```

We can see that the row that is returned is not caught in any of the filters!!
